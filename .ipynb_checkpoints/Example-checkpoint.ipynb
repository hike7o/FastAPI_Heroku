```python
def explain(customer_id:int):  
    explainer = shap.TreeExplainer(best_model)

    # client_index = test_set[test_set['SK_ID_CURR'] == 100038].index.item()
    X_shap = test_set.set_index('SK_ID_CURR')

    X_test_courant = X_shap.loc[[customer_id]]
    X_test_courant_array = X_test_courant.values.reshape(1, -1)

    shap_values_courant = explainer.shap_values(X_test_courant_array)
    exp_val = explainer.expected_value[1]
    return exp_val
# shap.decision_plot(explainer.expected_value[1], shap_values_courant[1],
#                    X_test_courant)
exp_val = explain(100057)
exp_val
```

```python
pickle.dump(exp_val, open('exp_value.pkl', 'wb'))
```

```python
def get_shap_data(customer_id):

    # save the response to API request
    response = requests.get("http://localhost:8000/shap_data/?SK_ID_CURR="+ str(customer_id))
    # convert from JSON format to Python dict
    content = json.loads(response.content)
    # convert data to pd.Series
    cust_shap_data = content
    return cust_shap_data

shap_data_customer = get_shap_data(100057)
np.array(shap_data_customer['1'])
```
